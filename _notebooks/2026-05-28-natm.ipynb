## Summary

We presented the OCS Capstone Platform — a searchable grid of 43 student projects backed by a Spring Boot API — to an audience of parents, teachers, and students. Visitors interacted with the live site, asked questions about the database and search functionality, and left written feedback. The event confirmed what works and surfaced two clear areas to improve.

![Team at Night at the Museum 2026]({{site.baseurl}}/images/natm_2026.png)

## What We Showed

| Feature | What visitors saw |
|---|---|
| Project grid | 43 capstone cards, live search and filter |
| CRUD modals | Adding and editing a project card in real time |
| Database | Spring Boot backend storing project data, visible in the UI |
| Games directory | 21 educational games with expandable learning drawers |
| Leaderboard | Top-10 projects ranked by upvotes, auto-refreshing |

## Feedback

**What landed well:**
- The site looked professional and was easy to navigate without explanation
- Storing projects in a database instead of hardcoded files made sense to non-technical visitors
- The search box was the feature people used most on their own

**What to improve:**
- Editing permissions need tightening — visitors noticed a user could accidentally edit someone else's card
- Some sections should be optional so projects that don't have a game aren't forced to show an empty slot

## Reflection

Presenting to a non-technical audience was harder than I expected. Explaining *why* a database matters — instead of just showing that it works — required a different kind of thinking than writing the code did. The feedback about editing permissions was something we had discussed internally but deprioritized; hearing it from a visitor made it concrete.

The part I found most useful: watching people use the site without any guidance. Where they clicked first, where they got confused, and what they asked about told me more than any code review would have.
