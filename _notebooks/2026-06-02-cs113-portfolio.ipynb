## Requirements Met — Summary

All 12 CS 113 college-credit requirements are satisfied. Here is a plain-language account of each one.

**Backend Java Spring Boot with JPA/Hibernate** — The backend (`nikhilsna/spring`) is a full Spring Boot application. Every database entity uses `@Entity`, `@OneToMany`, and `@ManyToMany` annotations backed by Hibernate. JPA repositories handle all persistence.

**3+ data structures appropriately applied** — Seven are present in production code: HashSet (WebSocket session tracking), HashMap (per-user stats), Stack and Queue (RPN calculator), Graph with BFS (assignment traversal), List (submission records), and Set (role management).

**2+ algorithms with complexity analysis** — Six are demonstrated: Comparator-based sorting in `RankingsController`, merge sort and bubble sort with empirical Big-O timing in `bigO.ipynb`, binary search and linear filtering in `lstfltng.ipynb`, BCrypt hashing in `MvcConfig`, and BFS graph traversal with O(V+E) analysis in `AssignmentGraphService`.

**Complete OOP design (all four pillars)** — Encapsulation via private fields and `@JsonIgnore`, Inheritance via `Person extends Submitter`, Polymorphism via three `@Override` methods in `WebSocketHandler`, and Abstraction via two abstract base classes (`Submitter`, `User`) and one interface (`AssignmentJpaRepository`).

**RESTful API with proper HTTP methods and error handling** — Over 100 endpoints across 15+ `@RestController` classes, each returning `ResponseEntity<>` with appropriate HTTP status codes.

**Database integration with entity relationships** — `@OneToMany`, `@ManyToMany`, and `@JoinTable` annotations on `Person` and `Assignment` define a normalized relational schema persisted to MySQL.

**JUnit tests** — 9 test files, 18 `@Test` methods using `@SpringBootTest` and `@MockBean` for integration-level coverage.

**Docker deployment with docker-compose** — `Dockerfile` uses `eclipse-temurin:21` multi-stage build; `docker-compose.yml` orchestrates the application on ports 8585 and 8589.

**Custom domain with DNS and nginx** — `nginx_spring_8585_8589.conf` proxies HTTPS traffic to the Spring backend with WebSocket upgrade support and Let's Encrypt TLS.

**CI/CD pipeline** — `.github/workflows/build.yml` runs `./mvnw -B test` against JDK 21 on every push via GitHub Actions.

**Blog portfolio documenting design and contributions** — This site (`nikhil-2026`) contains FRQ write-ups, algorithm notebooks, the SRP design pattern post, and this evidence summary.

**Project impact in production** — All features (Queue, Grade Requests, Capstone platform, Calendar, WebSocket, Bank/Rankings, ML grade predictor, BFS graph) are live on `opencodingsociety.com` and used by the class.
