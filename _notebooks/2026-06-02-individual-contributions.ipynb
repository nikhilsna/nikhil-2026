## Reflection

The part that required the most judgment was the controller design — specifically deciding how to handle duplicate titles (409 vs silently overwrite) and how to expose the endpoints publicly while keeping write operations behind auth. Getting that security config right meant the frontend grid could load for anyone without a login, while the add/edit/delete actions still required authentication.

The most visible result: before the CRUD system, adding a project meant editing a Jekyll file and waiting for a deploy. After it, anyone with an account can do it from the browser in under a minute.
