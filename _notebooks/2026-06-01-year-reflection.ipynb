```python
import matplotlib.pyplot as plt
import numpy as np

milestones = [
    ("Sep '24", "Python &\nJS Basics",       "#FF6B6B"),
    ("Oct '24", "Data Structures\nSprint",   "#FF9F43"),
    ("Nov '24", "Practice Exam 1\n81%",      "#FECA57"),
    ("Feb '25", "ML & Pandas\nSprint",       "#1DD1A1"),
    ("Apr '25", "Algorithms\n& Big O",        "#A29BFE"),
    ("Jun '25", "End of\nYear 1",            "#FD79A8"),
    ("Jan '26", "FRQ Practice\nBegins",      "#48DBFB"),
    ("Mar '26", "FRQ Blitz\n(4 FRQs)",      "#6C5CE7"),
    ("May '26", "AP Exam +\nCapstone",       "#00CEC9"),
]

fig, ax = plt.subplots(figsize=(16, 4.5))
fig.patch.set_facecolor('#0f0c29')
ax.set_facecolor('#16213e')

xs = np.linspace(0.07, 0.93, len(milestones))
ax.plot([0.04, 0.96], [0.5, 0.5], color='#4a4a6a', linewidth=2.5, zorder=1)

for i, ((date, label, color), x) in enumerate(zip(milestones, xs)):
    y_box = 0.80 if i % 2 == 0 else 0.16
    ax.plot([x, x], [0.5, y_box], color=color, linewidth=1.2, alpha=0.55, zorder=2, linestyle='--')
    ax.scatter(x, 0.5, s=280, color=color, zorder=4, edgecolors='white', linewidths=1.5)
    ax.text(x, y_box, f"{date}\n{label}", ha='center', va='center',
            fontsize=7.8, fontweight='bold', color='white',
            bbox=dict(boxstyle='round,pad=0.45', facecolor=color, alpha=0.88, edgecolor='none'))

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
ax.set_title('CS Journey — Key Milestones   (Sep 2024 → May 2026)',
             fontsize=13, fontweight='bold', color='white', pad=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months  = ["Sep\n'24","Oct\n'24","Nov\n'24","Feb\n'25","Mar\n'25",
           "Apr\n'25","May\n'25","Jun\n'25","Jan\n'26","Feb\n'26",
           "Mar\n'26","May\n'26"]
counts  = [6, 10, 1, 5, 5, 9, 1, 3, 2, 1, 5, 3]
clrs    = ['#FF6B6B','#FF9F43','#FECA57','#1DD1A1','#48DBFB',
           '#A29BFE','#FD79A8','#FDCB6E','#6C5CE7','#00CEC9','#E17055','#74B9FF']

fig, ax = plt.subplots(figsize=(14, 5))
fig.patch.set_facecolor('#0f0c29')
ax.set_facecolor('#16213e')

bars = ax.bar(months, counts, color=clrs, edgecolor='none', width=0.62)
for bar, c in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.18,
            str(c), ha='center', fontsize=11, fontweight='bold', color='white')

ax.set_ylabel('Notebooks / Assignments', color='#b2bec3', fontsize=10)
ax.set_title('Monthly Output Volume — Full Year', fontsize=13, fontweight='bold', color='white', pad=12)
ax.tick_params(colors='#b2bec3', labelsize=9)
for spine in ax.spines.values():
    spine.set_color('#2d3436')
ax.set_ylim(0, 13)
ax.grid(axis='y', color='#2d3436', linewidth=0.8)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories  = ['Java / OOP', 'Algorithms\n& Big O', 'Data\nStructures',
                'Web Dev\n(JS/HTML)', 'ML /\nPandas', 'Problem\nSolving']
N           = len(categories)
start_vals  = [2, 1, 2, 3, 1, 2]
end_vals    = [5, 4, 4, 4, 3, 5]

angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles   += angles[:1]
start_vals += start_vals[:1]
end_vals   += end_vals[:1]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
fig.patch.set_facecolor('#0f0c29')
ax.set_facecolor('#16213e')

ax.plot(angles, start_vals, 'o-', linewidth=2, color='#74b9ff', label='Start of Year', alpha=0.9)
ax.fill(angles, start_vals, alpha=0.12, color='#74b9ff')
ax.plot(angles, end_vals,   'o-', linewidth=2.5, color='#fd79a8', label='End of Year',   alpha=0.9)
ax.fill(angles, end_vals,   alpha=0.18, color='#fd79a8')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10, color='white')
ax.set_ylim(0, 5)
ax.set_yticks([1,2,3,4,5])
ax.set_yticklabels(['1','2','3','4','5'], fontsize=7, color='#636e72')
ax.spines['polar'].set_color('#2d3436')
ax.grid(color='#2d3436', linewidth=0.8)
ax.set_title('Skills Growth — Start vs End of Year', fontsize=13, fontweight='bold',
             color='white', pad=24)
ax.legend(loc='upper right', bbox_to_anchor=(1.38, 1.15),
          facecolor='#16213e', edgecolor='#636e72', labelcolor='white')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('#0f0c29')

# ── Left: Score Progression ──────────────────────────────────────────────
ax1 = axes[0]
ax1.set_facecolor('#16213e')
labels  = ["Practice Exam 1\n(Nov '24)", "Practice Exam 2\n(Feb '26)", "Progress\nChecks Avg"]
scores  = [81, 76, 88]
dot_clr = ['#48DBFB', '#1DD1A1', '#FDCB6E']

ax1.plot(range(3), scores, '-', linewidth=2.5, color='#A29BFE', zorder=2)
ax1.fill_between(range(3), scores, 60, alpha=0.08, color='#A29BFE')
for i, (s, c) in enumerate(zip(scores, dot_clr)):
    ax1.scatter(i, s, s=220, color=c, zorder=4, edgecolors='white', linewidths=1.5)
    ax1.annotate(f'{s}%', (i, s), textcoords='offset points', xytext=(0,13),
                 ha='center', fontsize=12, fontweight='bold', color=c)
ax1.axhline(75, color='#FF6B6B', linestyle='--', linewidth=1.2, alpha=0.7, label='75% target')
ax1.axhline(90, color='#1DD1A1', linestyle='--', linewidth=1.2, alpha=0.7, label='90% target')
ax1.set_xticks(range(3))
ax1.set_xticklabels(labels, fontsize=8.5, color='white')
ax1.set_ylim(60, 105)
ax1.set_ylabel('Score %', color='#b2bec3')
ax1.set_title('AP CSA Score Progression', fontsize=12, fontweight='bold', color='white', pad=10)
ax1.tick_params(colors='#b2bec3')
for sp in ax1.spines.values(): sp.set_color('#2d3436')
ax1.grid(color='#2d3436', linewidth=0.8)
ax1.legend(facecolor='#16213e', edgecolor='#636e72', labelcolor='white', fontsize=9)

# ── Right: Unit Breakdown ────────────────────────────────────────────────
ax2 = axes[1]
ax2.set_facecolor('#16213e')
units  = ['U1: Objects &\nMethods','U2: Selection &\nIteration','U3: Class\nCreation','U4: Data\nCollections']
uscores = [75, 85, 71, 81]
uclrs   = ['#48DBFB','#1DD1A1','#FF6B6B','#FDCB6E']
bars = ax2.barh(units, uscores, color=uclrs, edgecolor='none', height=0.5)
for bar, s in zip(bars, uscores):
    ax2.text(bar.get_width() + 0.8, bar.get_y() + bar.get_height()/2,
             f'{s}%', va='center', fontsize=11, fontweight='bold', color='white')
ax2.axvline(75, color='#FF6B6B', linestyle='--', linewidth=1.2, alpha=0.6)
ax2.set_xlim(0, 108)
ax2.set_title('Unit Performance (Practice Exam 1)', fontsize=12, fontweight='bold', color='white', pad=10)
ax2.tick_params(colors='#b2bec3')
for sp in ax2.spines.values(): sp.set_color('#2d3436')
ax2.grid(axis='x', color='#2d3436', linewidth=0.8)
ax2.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

frqs       = ['FRQ 1\nMethods &\nControl Flow', 'FRQ 2\nClasses', 'FRQ 3\nArray /\nArrayList', 'FRQ 4\n2D Arrays']
difficulty = [2, 3, 3, 5]
confidence = [5, 4, 4, 2]

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#0f0c29')
ax.set_facecolor('#16213e')

x = np.arange(len(frqs))
w = 0.35
bd = ax.bar(x - w/2, difficulty, w, label='Difficulty', color='#FF6B6B', edgecolor='none')
bc = ax.bar(x + w/2, confidence, w, label='Confidence', color='#48DBFB', edgecolor='none')
for bar in list(bd) + list(bc):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            str(int(bar.get_height())), ha='center', fontsize=12, fontweight='bold', color='white')

ax.axvspan(2.55, 3.45, alpha=0.07, color='#FF6B6B')
ax.text(3, 5.75, '⚠  Hardest', ha='center', fontsize=9.5, color='#FF6B6B', fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(frqs, fontsize=9, color='white')
ax.set_ylim(0, 6.5)
ax.set_ylabel('Rating (1–5)', color='#b2bec3')
ax.set_title('FRQ Practice Log — Difficulty vs Confidence', fontsize=13, fontweight='bold', color='white', pad=12)
ax.tick_params(colors='#b2bec3')
for sp in ax.spines.values(): sp.set_color('#2d3436')
ax.grid(axis='y', color='#2d3436', linewidth=0.8)
ax.legend(facecolor='#16213e', edgecolor='#636e72', labelcolor='white', fontsize=10)
plt.tight_layout()
plt.show()
```

  <p>Two years of computer science taught me more than just syntax. Year 1 was about building foundations — Python, JavaScript, data structures, and machine learning basics.
