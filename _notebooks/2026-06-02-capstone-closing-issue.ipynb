## For the Teacher

The table below maps every task that used to require manual file editing to the automated equivalent now built into the platform. If you find yourself doing something from the left column, the right column is where to go instead.

<br>

| Instead of... | Now you can... |
|---|---|
| Editing Jekyll files to publish a student project | Click **+** on `/capstone/` — fill out the form, card appears live |
| Scrolling through 43 cards to find something | Type in the search box — results filter in real time |
| Asking students what tech stack they used | Hover any project card — the tooltip lists the full stack |
| Manually tracking which projects students care about | Check `/capstone/leaderboard/` — ranked by upvote count, updates live |
| Explaining the platform to new students each semester | Send them to `/capstone-home-documentation/` — timestamped video walkthrough |

## The 5 Files That Run Everything

You do not need to understand the full codebase to maintain this platform. These 5 files are the only ones you should ever need to touch for day-to-day updates. Each row tells you exactly what to change and where.

<br>

| # | File | What it controls | How to edit it |
|:---:|---|---|---|
| **1** | `_posts/capstone/2026-02-09-capstone_home.md` | Project grid, CRUD modals, search & filter | Add cards inside `div#capstone-grid`; update `linkMap` for new URLs |
| **2** | `_data/capstone_card_tech.yml` | Hover tooltips that show each project's tech stack | Add a new entry: `"Project Title": ["Tech1", "Tech2"]` |
| **3** | `_projects/.makeprojects` | Which games are registered and included in the build | Add one line per new game: `games/your-game-name` |
| **4** | `_posts/capstone/2026-05-31-games-directory.md` | Games showcase page and learning drawers | Update the `GAME_DATA` JS object at the top of the file |
| **5** | `_posts/capstone/2026-04-13-capstone_home_documentation.md` | Video how-to guide and timestamp jump buttons | Swap the MP4 src; update each `seekCapstoneVideo(N)` call with new timestamps |
