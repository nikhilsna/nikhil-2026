      <p style="margin-top:.5rem;font-family:'Fira Code',monospace;font-size:12px;color:#5b21b6">push(3) → [3]<br>push(5) → [3,5]<br>pop() → 5, stack=[3]</p>

      <h4>&#x1F69B; Queue — FIFO</h4>
      <p><strong>First In, First Out.</strong> Like a line at a store — first person in is first out. Enqueue adds to back. Dequeue removes from front.</p>
      <p style="margin-top:.5rem;font-family:'Fira Code',monospace;font-size:12px;color:#1d4ed8">enqueue(3) → [3]<br>enqueue(5) → [3,5]<br>dequeue() → 3, queue=[5]</p>
    </div>

  <div class="sechdr"><span class="secnum">2</span><h2>How RPN uses both structures</h2></div>
  <p>Reverse Polish Notation puts operators <em>after</em> their operands. <span style="font-family:'Fira Code',monospace;background:#dbeafe;color:#1e40af;padding:1px 6px;border-radius:4px;font-size:12px">3 4 + 2 *</span> means (3 + 4) × 2 = 14.</p>
  <p>The algorithm has two phases:</p>
  <div class="code"><span class="cm">// Phase 1: tokenize input into a Queue (left to right)</span>
Queue&lt;<span class="cls">String</span>&gt; tokenQueue = <span class="kw">new</span> LinkedList&lt;&gt;();
<span class="kw">for</span> (<span class="cls">String</span> token : expression.<span class="fn">split</span>(<span class="str">" "</span>)) {
    tokenQueue.<span class="fn">enqueue</span>(token);   <span class="cm">// FIFO — preserve left-to-right order</span>
}

<span class="cm">// Phase 2: evaluate using a Stack</span>
Stack&lt;<span class="cls">Double</span>&gt; operandStack = <span class="kw">new</span> Stack&lt;&gt;();
<span class="kw">while</span> (!tokenQueue.<span class="fn">isEmpty</span>()) {
    <span class="cls">String</span> token = tokenQueue.<span class="fn">dequeue</span>();  <span class="cm">// pull next token</span>
    <span class="kw">if</span> (<span class="fn">isNumber</span>(token)) {
        operandStack.<span class="fn">push</span>(<span class="cls">Double</span>.<span class="fn">parseDouble</span>(token)); <span class="cm">// push operand</span>
    } <span class="kw">else</span> {
        <span class="cm">// operator: pop TWO operands, compute, push result</span>
        <span class="cls">double</span> b = operandStack.<span class="fn">pop</span>();  <span class="cm">// right operand (popped first)</span>
        <span class="cls">double</span> a = operandStack.<span class="fn">pop</span>();  <span class="cm">// left operand</span>
        operandStack.<span class="fn">push</span>(<span class="fn">compute</span>(a, token, b));
    }
}
<span class="cm">// final answer sits alone on the stack</span>
<span class="kw">return</span> operandStack.<span class="fn">pop</span>();</div>

function nextStep(){
  if(stepIndex>=tokens.length){
    document.getElementById('step-log').textContent='Done! Final answer: '+stack[0];
    document.getElementById('step-log').className='step-log active';
    document.getElementById('next-btn').disabled=true;
    return;
  }
  var t=tokens[stepIndex];
  var log='';
  if(!isNaN(t)){
    stack.push(parseFloat(t));
    log='Token "'+t+'" is a number → push onto stack. Stack: ['+stack.join(', ')+']';
  } else {
    var b=stack.pop(), a=stack.pop(), res;
    if(t==='+') res=a+b;
    else if(t==='-') res=a-b;
    else if(t==='*') res=a*b;
    else if(t==='/') res=a/b;
    stack.push(res);
    log='Token "'+t+'" is operator → pop '+b+' (b) and '+a+' (a) → '+a+t+b+'='+res+' → push '+res+'. Stack: ['+stack.join(', ')+']';
  }
  stepIndex++;
  document.getElementById('step-log').textContent=log;
  document.getElementById('step-log').className='step-log active';
  renderViz();
}

function resetViz(){
  stepIndex=0; stack=[];
  document.getElementById('step-log').textContent='Press Next Step to begin';
  document.getElementById('step-log').className='step-log';
  document.getElementById('next-btn').disabled=false;

function evalRPN(){
  var expr=document.getElementById('rpn-input').value.trim();
  var rb=document.getElementById('result-box');
  if(!expr){rb.className='result-box err show';rb.textContent='Enter an expression first.';return;}
  var toks=expr.split(/\s+/);
  var st=[];
  try{
    toks.forEach(function(t){
      if(!isNaN(t)){st.push(parseFloat(t));}
      else{
        if(st.length<2) throw new Error('not enough operands');
        var b=st.pop(),a=st.pop(),r;
        if(t==='+') r=a+b;
        else if(t==='-') r=a-b;
        else if(t==='*') r=a*b;
        else if(t==='/') r=a/b;
        else throw new Error('unknown operator: '+t);
        st.push(r);
      }
    });
    if(st.length!==1) throw new Error('invalid expression');
    rb.className='result-box show';
    rb.textContent='Result: '+st[0];
    if(!calcdone){addXP(100);calcdone=true;}
  } catch(e){
    rb.className='result-box err show';
    rb.textContent='Error: '+e.message;
  }
}
