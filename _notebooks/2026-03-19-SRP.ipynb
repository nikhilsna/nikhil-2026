  <div class="sechdr"><span class="secnum">1</span><h2>What is SRP — and why do comments matter?</h2></div>
  <span class="badge b-purple">SOLID · S</span>
  <span class="badge b-blue">Design Principle</span>
  <p>The <strong>Single Responsibility Principle</strong> says every class should have <em>one job and one job only</em>. A class that does too many things becomes impossible to maintain — changing one part breaks another.</p>
  <p>But SRP doesn't stop at how you split your classes. <strong>Comments are the contract</strong> — they tell future readers (and your teacher during live review) exactly what each class is responsible for, and just as importantly, what it deliberately <em>doesn't</em> do.</p>
  <div class="tip">&#128161; Think of a good SRP comment like a job description. It names the role, lists the duties, and makes clear what's out of scope.</div>
  <p>In the AutoTriage backend, four Java classes each own a single layer:</p>
  <div style="display:grid;grid-template-columns:1fr 1fr;gap:8px;margin:.75rem 0">

 */</span>

<div class="card">
  <div class="sechdr"><span class="secnum">4</span><h2>Inline comments that explain the boundary</h2></div>
  <p>Inline comments should explain <em>decisions</em>, not describe syntax. From the real <span style="font-family:'Fira Code',monospace;background:#1e1a35;color:#c4b5fd;padding:1px 6px;border-radius:4px;font-size:12px">ActivityService.java</span>:</p>
  <div class="code"><span class="cm">// Constructor injection keeps this testable —</span>
<span class="cm">// swap a mock repo in unit tests without touching this class.</span>
<span class="kw">private final</span> ActivityRepository repository;

<span class="cm">// Rejects empty file lists before they reach the repository.</span>
<span class="cm">// Business rule enforced here — not in the model, not in the controller.</span>
<span class="kw">public void</span> <span class="fn">recordSession</span>(ActivityEvent event) {
    <span class="kw">if</span> (event.getFilesModified() == <span class="kw">null</span>) {
        <span class="kw">throw new</span> IllegalArgumentException(<span class="str">"..."</span>);
    }
    repository.<span class="fn">save</span>(event);
}

<span class="cm">// "Inactive" is a business concept — INACTIVE_THRESHOLD_HOURS</span>
<span class="cm">// lives here so Mortensen can tune it without hunting magic numbers.</span>
<span class="kw">private static final int</span> INACTIVE_THRESHOLD_HOURS = <span class="num">48</span>;</div>
  <p>Each comment answers: <em>why does this line exist, and why does it live in this class?</em></p>

  if (dragdone) return;
  var tags = document.querySelectorAll('.tag[data-id]');
  var correct = 0;
  tags.forEach(function(tag) {
    var parent = tag.parentElement;
    if (parent && parent.id === tag.dataset.answer) {
      correct++;
      parent.classList.add('z-ok');
    } else if (parent && parent.classList.contains('drag-zone')) {
      parent.classList.add('z-no');
    }
  });
  var result = document.getElementById('drag-result');
  if (correct === 5) {
    result.textContent = '5 / 5 correct!';
    result.style.color = '#6ee7b7';
    addXP(200); dragdone = true;
  } else if (correct >= 3) {
    result.textContent = correct + ' / 5 correct — close!';
    result.style.color = '#fde68a';
    addXP(80); dragdone = true;
  } else {
    result.textContent = correct + ' / 5 — try again after reset';
    result.style.color = '#fca5a5';
  }
}

function resetDrag() {
  if (dragdone) return;
  var bank = document.getElementById('bank');
  document.querySelectorAll('.tag[data-id]').forEach(function(t){ bank.appendChild(t); });
  document.querySelectorAll('.drag-zone').forEach(function(z){
    z.classList.remove('z-ok','z-no','over');
  });
  document.getElementById('drag-result').textContent = '';
}
