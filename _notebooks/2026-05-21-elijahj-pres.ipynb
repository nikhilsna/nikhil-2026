## Before the Event

I expected a very interesting presentation on SafeBallot, and I expected Mr. Johnson to spend a lot of time on the cybersecurity aspect of his company.

---

## Curiosity Before Attending

I was most curious about how Mr. Johnson was able to receive funding for SafeBallot and how he was able to manage being CEO of his company.

---

## What I Saw

Elijah Johnson, SafeBallot booth.

---

## Most Interesting Thing I Saw

The most interesting thing I saw was Elijah using a Lamborghini in order to demonstrate tangible marketing to 5th graders.

---

## What I Learned

**People don't just buy products. People buy emotions.** If you want to successfully market a product, you must get down to the emotional cell.

---

## What Surprised Me

Elijah's team came in **second place** in the entrepreneurship competition to a massage chair, which I thought was surprising.

---

## What Inspired Me

Something that inspired me was Mr. Johnson's mention of **"No dream is too big."** When he wanted to unite 18 counties, everybody thought he was crazy — however, he only needed one **"yes"**, and that answer took him further than he could have ever imagined.

---

## Personal Connection

This connected to my personal interests because Mr. Johnson mentioned **relationship building**, which is something I'm very interested in. Sometimes, you need to be able to meet your clients at their level — for example, sports (a personal interest of mine) — before you can ease them into your product.

---

## Interaction

**Person or group I learned from:** Elijah Johnson

Elijah explained to me how he was able to use his prior entrepreneurial experience through many of his different businesses in his favor to manage SafeBallot and successfully develop and market it.

---

## Depth of Experience

This event changed my thinking because I assumed that the smaller relationships that people don't really think are important are actually the **most important drivers of change** behind what can make or break your product marketing.

---

## Connection to CS or Future Goals

This connects to my CS project because it mirrors the values of making a product **better, faster, and cheaper.** The final goal of our capstone project once it is finished is to make displaying projects better, faster, and more efficient.

---

## Final Reflection

This event was meaningful to me because it reaffirmed my ideas of success: Not pursuing to be the best possible "XYZ", but pushing to **elevate yourself to the best you.**

---

## Biggest Takeaway

> **People don't buy products, they buy emotions. You need to be able to market the emotion of a product before selling it.**
