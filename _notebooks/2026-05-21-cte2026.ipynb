**TL;DR — Attended the CTE Expo at Del Norte on May 21. Saw robotics teams, a 70mph dune buggy from Abraxas HS, and STEM builds from multiple schools. Talked with Emmy from Abraxas about their build pipeline and how they handled real-world hardware constraints. Main lesson: great projects happen when solid code meets physical engineering. Biggest takeaway — start building before you feel ready.**

---

## Before the Event

I expected to see a few standard science fair poster boards and some basic classroom coding projects. I figured it would mostly be a casual showcase of standard high school elective work.

---

## Curiosity Before Attending

I was curious about how the other robotics teams designed their robots inside their own high school laboratories, and whether it was similar to what we have at Del Norte (laser cutter, dedicated robotics lab).

---

## What I Saw

I visited the DNHS robotics team's mobile arena setup, the computer science project tables in A101, the design mixed media booths, and the outdoor exhibition area where the different robots were racing each other.

---

## Most Interesting Thing I Saw

The most interesting thing was definitely the **dune buggy brought in from Abraxas High School.** Seeing a fully functional, low cost dune buggy built from a kit that didn't require consistent check-ups and had a top speed of **70 mph** was definitely very interesting.

---

## What I Learned

I learned about the **hardware integration required to make autonomous systems function reliably outdoors.** The robotics students explained how they have to calibrate their sensors and drivetrain to handle uneven pavement and bright sunlight compared to a flat indoor lab floor.

---

## What Surprised Me

I was surprised by the **sheer scale and collaborative nature** of the projects. Whole teams were working together, combining mechanical engineering, fabrication, and software development into physical builds.

---

## What Inspired Me

Seeing the passion and technical depth of the students presenting their work inspired me. I enjoyed seeing the dedication with which different students were working on their projects, **sacrificing sleep, food, and fun** in order to work on their project and benefit their team.

---

## Personal Connection

Going to this event directly connected to my personal interests because it demonstrated how students were able to **turn code into impactful projects.** While it can get boring to constantly write code and watch it execute on screen, it is a whole other realm to watch your code be executed in real life.

---

## Interaction

**Person or group I learned from:** Emmy (Abraxas High School)

She walked me through their building pipeline, showing me how they debugged their code and handled hardware limitations. They also explained the **cross-school collaboration** involved in bringing heavy-duty projects to the expo.

---

## Depth of Experience

It made me realize that **computer science and real world constraints and system design are mutually inclusive.** The best projects happen when robust software intersects with creative hardware and physical engineering to solve problems.

---

## Connection to CS or Future Goals

This connects to my AP Computer Science A class because it showed why **writing clean and efficient code actually matters.** In robotics and computer vision, bad logic does not just throw an error — it can also cause major setbacks for your entire workforce.

---

## Final Reflection

It was meaningful because it connected what I learn in class to something real. Seeing students my age build **working robots and AI systems** made me more excited about where computer science can actually take me.

---

## Biggest Takeaway

> **My biggest takeaway was that creating a successful project relies on one starting before they're ready. If you keep waiting till you have the right time, the right tools, the right ideas, you'll never build anything meaningful.**
